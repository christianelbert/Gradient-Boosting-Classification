```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import classification_report,roc_curve,plot_confusion_matrix, plot_roc_curve
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import AdaBoostClassifier

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('gabung.csv')
df.head()
```

```python
X = df.drop(columns=['target','track','artist'])
y = df['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=42)
```

### Chosen model

```python
ctb = CatBoostClassifier()
xgb = XGBClassifier()
lgb = LGBMClassifier()
ada = AdaBoostClassifier()
```

```python
def tuning(model, parameter, n_cv):
    grid = RandomizedSearchCV(model, parameter, cv = n_cv)
    grid.fit(X_train,y_train)
    best_param = grid.best_params_
    return best_param
```

```python
cat_param = {'depth': [4, 7, 10],
          'learning_rate' : [0.03, 0.1, 0.15],
         'l2_leaf_reg': [1,4,9],
         'iterations': [300]}

xgb_param = {"max_depth": [10,30,50],
              "min_child_weight" : [1,3,6],
              "n_estimators": [200],
              "learning_rate": [0.05, 0.1,0.16],}

lgb_param = {"max_depth": [25,50, 75],
              "learning_rate" : [0.01,0.05,0.1],
              "num_leaves": [300,900,1200],
              "n_estimators": [200]
             }
```

```python
lgb_best = tuning(lgb, lgb_param, 5)
```

```python
xgb_best = tuning(xgb, xgb_param, 5)
```

```python
cat_best = tuning(ctb, cat_param, 5)
```

```python
print(cat_best)
print(xgb_best)
print(lgb_best)
```

```python
ctb_opt = CatBoostClassifier(learning_rate=0.03,l2_leaf_reg=9,iterations=300,depth=7)
xgb_opt = XGBClassifier(n_estimators=200,min_child_weight=3,max_depth=10,learning_rate=0.05)
lgb_opt = LGBMClassifier(num_leaves=300,n_estimators=200,max_depth=50,learning_rate=0.01)
```

```python
ctb.fit(X_train,y_train)
xgb.fit(X_train,y_train)
lgb.fit(X_train,y_train)
```

```python
ctb_opt.fit(X_train,y_train)
xgb_opt.fit(X_train,y_train)
lgb_opt.fit(X_train,y_train)
```

```python
def model_score(model):
    return round(model.score(X_test,y_test)*100, 2)
```

```python
scoreNotune = pd.DataFrame({'Before tuning (train)':[ctb.score(X_train,y_train), xgb.score(X_train,y_train), lgb.score(X_train,y_train)]},index=['Catboost','XGBoost','LightGBM']).T
scoreTune = pd.DataFrame({'After tuning (train)':[ctb_opt.score(X_train,y_train),xgb_opt.score(X_train,y_train),lgb.score(X_train,y_train)]}, index=['Catboost','XGBoost','LightGBM']).T

train_score = pd.concat([scoreNotune,scoreTune])
```

```python
testscoreNotune = pd.DataFrame({'Before tuning (test)':[ctb.score(X_test,y_test), xgb.score(X_test,y_test), lgb.score(X_test,y_test)]},index=['Catboost','XGBoost','LightGBM']).T
testscoreTune = pd.DataFrame({'After tuning (test)':[ctb_opt.score(X_test,y_test),xgb_opt.score(X_test,y_test),lgb.score(X_test,y_test)]}, index=['Catboost','XGBoost','LightGBM']).T

test_score = pd.concat([testscoreNotune,testscoreTune])
```

```python
pd.concat([train_score,test_score])
```

```python
optimum = {
    'Catboost':ctb_opt,'XGBoost':xgb_opt,'LightGBM':lgb_opt
}

def classification(model):
    for key in list(optimum.keys()):
        print(f'Classfication report for {key}:\n {classification_report(y_test,optimum[key].predict(X_test))}') 

classification(optimum)
```

```python
keys = list(optimum.keys())

def confuse(model):
    for key in keys:
        print(f'Confusion matrix for {key}')
        plot_confusion_matrix(optimum[key],X_test,y_test,values_format='0.0f',)
        plt.show()

confuse(keys)
```

```python
plot_roc_curve(ctb_opt,X_test,y_test)
```

```python
plot_roc_curve(xgb_opt,X_test,y_test)
```

```python
plot_roc_curve(lgb_opt,X_test,y_test)
```

In predicting whether a track is a hit or not, false-negatives is a less of a concern, hence, precision will be the metric used to choose the best model.

In terms of precision, Catboost and XGBoost performance is similar. However, since Catboost can distinguish between hit and flop class slightly better (AUC score), Catboost will be the chosen model.
